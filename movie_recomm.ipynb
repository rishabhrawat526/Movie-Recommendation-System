```python
# FEATURE SELECTION
# genres
# id
# keywords
# title
# overview
# cast
# crew
```

```python
feat=['genres','id','keywords','title','overview','cast','crew']
for set_ in (movies):
    if set_ in feat:
        pass
    else:
        movies.drop(set_,axis=1,inplace=True)
  
```

```python
import ast
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i["name"])
    return l
```

```python

movies['genres'] = movies.genres.apply(lambda x: convert(x))
```

```python
movies.head(5)
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x:convert(x))
```

```python
def convert3(obj):
    c=0
    l=[]
    for i in ast.literal_eval(obj):
        if c != 3:
            l.append(i["name"])
            c+=1
        else:
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(lambda x:convert3(x))
```
